<font color='Blue'>We have so far made progress with the biased walker with probability distribution of $p = 0.8$ and hence $q = 1 - p = 0.2$.<br>
We now proceed with a reset for the walker when $p = 0.3$ i.e it returns straight away to starting point when it hits that $p$

#### Random walk in 1D<p>
Body is initially resting at $x = 0$ and it will move either in postive or negative $x$ direction and resets to $x=0$ again for some certain probability.<br>
    The probability continues to be biased with $p = 0.6$ and $q = 0.4$ with additional resetting behaviour

## N = 30

```python
from scipy import stats
import random
import matplotlib.pyplot as plt

def random_walk(n):

    x = 0
    # particle postioned at x = 0
    for i in range(n):
        step = round(random.random(),2)
        # steps being randomly assigned either +ve or -ve
        if step == 0.30:
            x = 0
        else:
            if step <= 0.40:
                x = x-1
            else:
                x = x+1
    return (x)


number_of_walk = 30000
walk_length = 30
position = []
# length of walk defines how much steps can body take

for i in range(number_of_walk):
    walk = random_walk(walk_length)
#     (walk, "Distance from origin = ", abs(walk))
    position.append(walk)

print(stats.describe(position))
```

```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


df = pd.DataFrame(position, columns=['position'])

fig,ax1 = plt.subplots(dpi=100)
sns.kdeplot(data=df,x='position',ax=ax1, color='g')

# ax1.set_xlim(-15,25)
ax2 = ax1.twinx()
# ax1.text(-14,0.11,s=r'$\mu \Rightarrow N(p-q)$ = 6',bbox=dict(facecolor='red', alpha=0.5))

sns.histplot(data=df, x='position',discrete=True,ax=ax2, color='k', alpha = 0.2)

plt.show()
```

\end{cases}

# N = 100

```python
# Symmetric Resetting

from scipy import stats
import random
import matplotlib.pyplot as plt

def random_walk(n):

    x = 0
    # particle postioned at x = 0
    for i in range(n):
        step = round(random.random(),2)
        # steps being randomly assigned either +ve or -ve
        if step <= 0.30:
            x = 0
        else:
            x = x+1
    return (x)


number_of_walk = 30000
walk_length = 100
position = []
# length of walk defines how much steps can body take

for i in range(number_of_walk):
    walk = random_walk(walk_length)
#     (walk, "Distance from origin = ", abs(walk))
    position.append(walk)

print(stats.describe(position))
```

```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.DataFrame(position, columns=['position'])

fig,ax1 = plt.subplots(dpi=100)
sns.kdeplot(data=df,x='position',ax=ax1, color='coral',alpha = 1)

ax2 = ax1.twinx()
# ax1.text(15,0.016,s=r'$\mu \Rightarrow N(p-q)$ = 1600',bbox=dict(facecolor='red', alpha=0.5))
sns.histplot(data=df, x='position',discrete=True,ax=ax2, color='royalblue', alpha = 0.2)
plt.title("N = 100", color= 'r')
plt.grid(axis='y')
plt.show()
```
