Binomial Distribution for a number of steps N. The probability of going going left or right is same here,
name p=q=0.5



## N = 50 

```python
from scipy.stats import binom
import seaborn as sb

binom.rvs(size=1000,n=50,p=0.5)

data_binom = binom.rvs(n=50,p=0.5,loc=0,size=1000)
ax = sb.distplot(data_binom,
                  kde=True,
                  color='red',
                  hist_kws={"linewidth": 5,'alpha':0.3})
ax.set(xlabel='n', ylabel=r'$P(n)$')
```

```python
'''Random walk in 1D
Body is initially resting at x = 0 and it will move
either in postive or negative x direction
'''


import random

def random_walk(n):

    x = 0
    # particle postioned at x = 0
    for i in range(n):
        step = random.choice([-1,1])
        # steps being randomly assigned either +ve or -ve

        if step == -1:
            x = x-1
        else:
            x = x+1
    return (x)


number_of_walk = 300000
walk_length = 50
position = []
# length of walk defines how much steps can body take

for i in range(number_of_walk):
    walk = random_walk(walk_length)
#     (walk, "Distance from origin = ", abs(walk))
    position.append(walk)


```

```python
from matplotlib import pyplot as plt

def fac(n):
    value = 1
    for i in range(2, n+1):
        value = value * i
    return value

def C(n, k, p):
    return fac(n)/(fac(k) * (fac(n-k))) * pow(p, k) * pow(1-p, n-k)

for N in [20]:
    X = []
    Y = []
    for K in range(1, 20):
        X.append(K)
        Y.append(C(N, K, 0.5))
    plt.plot(X, Y)
    plt.legend(("N = 20", "N = 30", "N = 50"))
plt.show()
```
