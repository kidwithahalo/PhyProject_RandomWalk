><h3>Walker in presence of constant resetting probability $0 \leq r < 1$

The position of $x_n$ evolves by rule as <br>
$
x_n = \begin{cases}
0, \hspace{30mm} \text{   with proba. r   }  \\
x_{n-1} + \eta_{n} , \hspace{15mm} \text{ with proba. 1 - r  } 
\end{cases}
$

<h3><font color='Blue'>Symmetric walkers with $r = 0.5$ for constant resetting

# N = 100

```python

import numpy as np
from scipy import stats
import random
import matplotlib.pyplot as plt

global x
x = 0 

def random_walk(n):
    x_1 = 0
    # particle postioned at x = 0
    for i in range(n):
        step = round(random.random(),2)
        # steps being randomly assigned either +ve or -ve
        if step <= 0.50:
            x_1 = x
        elif step > 0.50:
            zeta = np.random.choice([-1,1])
            if zeta == -1:
                x_1 = x_1 - 1 
            else:
                x_1 = x_1 + 1
    return (x_1)


number_of_walk = 100
walk_length = 100
position = []
# length of walk defines how much steps can body take

for i in range(number_of_walk):
    walk = random_walk(walk_length)
#     (walk, "Distance from origin = ", abs(walk))
    position.append(walk)

print(stats.describe(position))


```

```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.DataFrame(position, columns=['position'])

plt.style.use('seaborn-poster')
fig,ax1 = plt.subplots(dpi=200)
sns.kdeplot(data=df,x='position',linestyle='--',ax=ax1)


ax2 = ax1.twinx()
# ax1.text(15,0.016,s=r'$\mu \Rightarrow N(p-q)$ = 1600',bbox=dict(facecolor='red', alpha=0.5))
sns.histplot(data=df, x='position',discrete=True,ax=ax2, alpha = 0.6)
plt.title("N = 100", color= 'r')
plt.savefig('N=100.png')
plt.show()
```
