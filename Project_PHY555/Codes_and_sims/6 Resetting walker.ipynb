## Resetting Walker

><h4>Progress so far

<font color='Blue'>We have so far made progress with the biased walker with probability distribution of $p = 0.8$ and hence $q = 1 - p = 0.2$.<br>
We now proceed with a reset for the walker when $p = 0.3$ i.e it returns straight away to starting point when it hits that $p$ 
<hr>   

#### Random walk in 1D<p>
Body is initially resting at $x = 0$ and it will move either in postive or negative $x$ direction and resets to $x=0$ again for some certain probability.<br>
    The probability continues to be biased with $p = 0.6$ and $q = 0.4$ with additional resetting behaviour
