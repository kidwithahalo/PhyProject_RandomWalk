```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(-1, 2, 100)
y = np.exp(x)

plt.subplot(1,2,1)
plt.plot(x, y)
plt.xlabel('$x$')
plt.ylabel('$\exp(x)$')

plt.subplot(1,2,2)
plt.plot(x, -np.exp(-x))
plt.xlabel('$x$')
plt.ylabel('$-\exp(-x)$')

plt.show()
```
