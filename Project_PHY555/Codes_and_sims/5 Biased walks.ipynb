# N = 10

```python
'''Random walk in 1D
Body is initially resting at x = 0 and it will move
either in postive or negative x direction
'''


import random

def random_walk(n):

    x = 0
    # particle postioned at x = 0
    for i in range(n):
        step = random.random()
        # steps being randomly assigned either +ve or -ve

        if step <= 0.2:
            x = x-1
        else:
            x = x+1
    return (x)


number_of_walk = 300000
walk_length = 10
position = []
# length of walk defines how much steps can body take

for i in range(number_of_walk):
    walk = random_walk(walk_length)
#     (walk, "Distance from origin = ", abs(walk))
    position.append(walk)

print(stats.describe(position))

```

```python
position
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

x = np.arange(np.min(position), np.max(position), 0.001)

#create range of y-values that correspond to normal pdf with mean and sd 
y = norm.pdf(x,np.mean(position),np.std(position))

#define plot 
fig = plt.figure(figsize=(9,6), dpi = 300)
plt.plot(x,y, label = "Theoretical", linestyle = 'dashed', color = 'b')
plt.hist(position,20, density = True, fc = 'brown', alpha = 1, label = "Simulation")

#choose plot style and display the bell curve 
plt.style.use('seaborn-pastel')
plt.legend(fontsize = 20)
plt.title("N = 10")
plt.savefig("N_10_biashed.pdf")
plt.show()
```

# MSD

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(9,6))

walkers = 30000
path = []
steps = 100
for walker in range(walkers):
    walker_path=[0]
    position = 0
    for i in range(1, steps):
        move = np.random.random()
        if move <= 0.2:
            position += -1
        else:
            position += 1
        walker_path.append(position)
    path.append(walker_path)
path = np.array(path)

# returns a vector of MSDs from a given path
msd = [(path[:, i]**2).mean() for i in range(1, steps)]
# you could also put range(1, len(path[0]))
```

```python
step = np.arange(1,100,1)
msd_N = msd/step

plt.plot(step,msd_N)
plt.show()
```
